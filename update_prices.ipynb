# Update DeFacto prices
This notebook will load the JSON, increase each price by 5% and reformat to use comma and a dollar sign.

```python
import json, re

with open(r"c:\Users\user\Catalog\DeFacto.json", 'r', encoding='utf-8') as f:
    data = json.load(f)

print(f"Loaded {len(data)} items")
```

```python
# helper to parse price strings

def parse_price(p):
    if not isinstance(p, str):
        return None
    m = re.search(r"([0-9]+[.,]?[0-9]*)", p)
    if not m:
        return None
    return float(m.group(1).replace(',', '.'))

count=0
for item in data:
    num = parse_price(item.get('price'))
    if num is not None:
        new = num * 1.05
        item['price'] = f"{new:.2f}".replace('.', ',') + ' $'
        count += 1

print(f"Updated {count} prices")
```

```python
with open(r"c:\Users\user\Catalog\DeFacto.json", 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print("File overwritten with new prices")
```
